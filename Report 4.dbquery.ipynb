```sql
CREATE OR REPLACE TABLE default.olist_products_raw
USING DELTA AS SELECT * FROM read_files(
  '/Volumes/workspace/default/sales/olist_products_dataset.csv',
  format => 'csv', header => true, inferSchema => true
);

CREATE OR REPLACE TABLE default.category_translation
USING DELTA AS SELECT * FROM read_files(
  '/Volumes/workspace/default/sales/product_category_name_translation.csv',
  format => 'csv', header => true, inferSchema => true
);

CREATE OR REPLACE TABLE default.olist_products USING DELTA AS
SELECT p.*, t.product_category_name_english
FROM      default.olist_products_raw  p
LEFT JOIN default.category_translation t
       ON p.product_category_name = t.product_category_name;

SELECT COUNT(*) FROM default.olist_products;
```
