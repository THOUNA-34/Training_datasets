```sql
CREATE OR REPLACE TABLE default.olist_reviews
USING DELTA AS SELECT * FROM read_files(
  '/Volumes/workspace/default/sales/olist_order_reviews_dataset.csv',
  format => 'csv', header => true, inferSchema => true
);
CREATE OR REPLACE TABLE default.olist_sellers
USING DELTA AS SELECT * FROM read_files(
  '/Volumes/workspace/default/sales/olist_sellers_dataset.csv',
  format => 'csv', header => true, inferSchema => true
);
CREATE OR REPLACE TABLE default.olist_payments
USING DELTA AS SELECT * FROM read_files(
  '/Volumes/workspace/default/sales/olist_order_payments_dataset.csv',
  format => 'csv', header => true, inferSchema => true
);
SELECT 'reviews' AS tbl,COUNT(*) AS rows FROM default.olist_reviews  UNION ALL
SELECT 'sellers' AS tbl,COUNT(*) AS rows FROM default.olist_sellers  UNION ALL
SELECT 'payments',   COUNT(*) AS rows FROM default.olist_payments;
```
