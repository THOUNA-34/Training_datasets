```sql
CREATE OR REPLACE TABLE default.master_analytics AS
SELECT
  o.order_id,
  o.customer_id,
  o.order_ts,
  o.fulfilment_days,
  o.order_status,

  c.customer_city,
  c.customer_state,

  oi.product_id,
  oi.seller_id,
  oi.price,
  oi.freight_value,

  p.product_category_name_english AS category,

  r.review_score,

  s.seller_city,
  s.seller_state,

  pay.payment_type,
  pay.payment_value

FROM default.clean_orders o
JOIN default.olist_customers c
  ON o.customer_id = c.customer_id
JOIN default.olist_order_items oi
  ON o.order_id = oi.order_id
JOIN default.olist_products p
  ON oi.product_id = p.product_id
LEFT JOIN default.olist_reviews r
  ON o.order_id = r.order_id
LEFT JOIN default.olist_sellers s
  ON oi.seller_id = s.seller_id
LEFT JOIN default.olist_payments pay
  ON o.order_id = pay.order_id;

SELECT COUNT(*) FROM default.master_analytics;
```
