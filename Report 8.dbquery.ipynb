```sql
CREATE OR REPLACE TABLE default.olist_products AS
SELECT
  p.*,
  t.product_category_name_english
FROM default.olist_products_raw p
LEFT JOIN default.category_translation t
ON p.product_category_name = t.product_category_name;

SELECT product_category_name, product_category_name_english
FROM default.olist_products
LIMIT 10;
```
