```sql
CREATE OR REPLACE VIEW default.vw_delivery_performance AS
SELECT
  order_status,
  COUNT(*) AS total_orders,
  ROUND(AVG(CAST(review_score AS INT)), 2) AS avg_review,
  ROUND(AVG(fulfilment_days), 1) AS avg_days
FROM default.master_analytics
WHERE review_score IS NOT NULL
GROUP BY order_status;

CREATE OR REPLACE VIEW default.vw_monthly_revenue AS
SELECT
  DATE_FORMAT(order_ts, 'yyyy-MM') AS month,
  ROUND(SUM(payment_value), 2) AS revenue,
  COUNT(DISTINCT order_id) AS orders
FROM default.master_analytics
WHERE payment_value IS NOT NULL
GROUP BY DATE_FORMAT(order_ts, 'yyyy-MM');

CREATE OR REPLACE VIEW default.vw_seller_scorecard AS
SELECT
  seller_id,
  seller_city,
  seller_state,
  COUNT(DISTINCT order_id) AS total_orders,
  ROUND(SUM(payment_value), 2) AS total_revenue,
  ROUND(AVG(review_score), 2) AS avg_rating,
  ROUND(AVG(fulfilment_days), 1) AS avg_delivery_days
FROM default.master_analytics
WHERE seller_id IS NOT NULL
GROUP BY seller_id, seller_city, seller_state;


SHOW VIEWS IN default;

SELECT * FROM default.vw_monthly_revenue LIMIT 5;
```
